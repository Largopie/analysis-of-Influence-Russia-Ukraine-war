# 전쟁으로 인한 러시아군 사망률과 전세계 수천명단위 연평균 사망자수 비교

```python
# 러시아군 사망자수
러시아군사망자수 = pd.read_csv('data/전쟁중_러시아군_사망자수.csv', sep = ',', engine = 'python',
                   names = ['일자','사망자수'], skiprows=1, index_col=0)
```

```python
# 우크라이나 민간인 사망자수
우크라이나민간인사상자수 = pd.read_csv("data/우크라이나민간인사상자.csv", sep = ',', engine = 'python',
                   names = ['일자','사망자','부상자'], skiprows=1, index_col=0, encoding='cp949')
```


### 총 러시아군 사망자 수 증가 그래프

```python
df = 러시아군사망자수

fig = px.bar(df, x=df.index, y=df.사망자수, title = '러시아군 사망자 증가')
fig.update_layout(plot_bgcolor='#eee', title_font_size=30, paper_bgcolor='#fff', title_x=0.5)
fig.show()
```

### 우크라이나 민간인 사상자 수

```python
df1 = 우크라이나민간인사상자수

fig = px.line(df1, x=df1.index, y=df1.columns,
              title='우크라이나 민간인 사상자 증가')
fig.update_layout(plot_bgcolor='#eee', title_font_size=30, paper_bgcolor='#fff', 
                  title_x=0.5, yaxis_title="사상자 수", legend_title="")
fig['data'][0]['line']['color']='rgb(255, 182, 193)'
fig['data'][1]['line']['color']='rgb(100, 204, 204)'
fig.show()
```

### 총 러시아군 사망자 수

```python
총러시아군사망자수 = 러시아군사망자수.max()
총러시아군사망자수
```

### 전세계 2022년 수천명 단위 사망자수


```python
# 세계 연평균 수천명 단위 사망자 수 [ 수천명 : 999,999명]
세계연평균사망자수 = pd.read_csv('data/전세계_수천명단위_연평균_사망자수.csv',sep = ',', engine = 'python',names = ['Date','Count'], skiprows=1)
세계연평균사망자수
```

```python
세계올해평균사망자수 = 세계연평균사망자수[세계연평균사망자수.Date==2022].Count
세계올해평균사망자수
```

```python
세계올해평균사망자수['러시아군사망자수'] = 총러시아군사망자수.사망자수;
세계올해평균사망자수.index = ['수천단위_올해_평균_사망자수','총_러시아군_사망자수']
사망자수비교 = 세계올해평균사망자수
사망자수비교
```

```python
fig = px.bar(총러시아군사망자수, x=사망자수비교.index, y=사망자수비교.values, 
             title='총 러시아군 사망자 수와 수천단위 올해 평균 사망자수 비교')
fig.update_layout(plot_bgcolor='#eee', title_font_size=30, paper_bgcolor='#fff', 
                  title_x=0.5, yaxis_title="", width=800, height= 500, xaxis_title="")
fig.update_traces(width=0.3)
fig.show()
```

올해의 세계 평균 사망자수(999,999명 평균)와 총 러시아군의 사망자 수를 비교하였을 때 <br>
러시아군의 사망자수가 더 많은 것을 그래프를 통해 확인할 수 있었다. <br>
<br>
물론 999,999명 단위 평균 사망자수와 비교하였다는 한계점이 있지만 <br>
이 값과 비교했다고 하더라도 결코 적은 값이 아니라는 것을 알 수 있다.<br>
<br>
러시아군 뿐만 아니라 우크라이나군, 해외에서 파병된 용병, 민간인 등 다른 참전자들의 사망자 수를 다 더하게 된다면<br>
현재 러시아군 사망자수의 3배정도가 될 것으로 예측이 된다.<br>
